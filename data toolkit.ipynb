3. Explain the following termsg
M The difference in np.array, np.asarray and np.asanyarrayX
M The difference between Deep copy and shallow copyX""


1. Difference between np.array, np.asarray, and np.asanyarray
np.array:

Creates a new ndarray object.
Always copies the data unless copy=False is explicitly provided, in which case it may still copy if the input is not already an ndarray.
Converts input into an array, allowing specification of dtype and other options.
np.asarray:

Converts the input to an array, but does not copy the data if the input is already an ndarray.
It is more efficient than np.array when working with arrays since it avoids redundant copying.
Preserves the dtype of the input.
np.asanyarray:

Similar to np.asarray, but it preserves subclasses of ndarray (e.g., matrix or masked_array).
This is useful when you want to ensure the output is at least an array, but you don’t want to lose specific subclass behavior.


2. Difference between Deep Copy and Shallow Copy
Shallow Copy:

Copies only the top-level structure of the object, but not the nested objects (sub-elements or references within).
Changes made to nested objects (like lists inside lists) in the copied object will affect the original object since they share references.
Example: copy.copy() performs a shallow copy.
Deep Copy:

Copies the object along with all the nested objects it contains. The copy is independent of the original.
Changes made in the copied object, including changes to nested objects, do not affect the original object.
Example: copy.deepcopy() performs a deep copy.

```python
"""5.Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from arrayX"""

import numpy as np

random_array = np.random.randint(1, 11, (5, 6))

even_integers = random_array[random_array % 2 == 0]

odd_integers = random_array[random_array % 2 != 0]

print("Original 5x6 array with random integers:\n", random_array)
print("\nEven integers from the array:\n", even_integers)
print("\nOdd integers from the array:\n", odd_integers)
```

```python
"""15.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
K§g Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:
KÃg Set the title of the plot as 'Histogram with PDF Overlay'.
Kçg Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Timp."""


#1. Create a Time-Series Dataset

import pandas as pd
import numpy as np

date_range = pd.date_range(start='2021-01-01', periods=100, freq='D')

data = {
    'Date': date_range,
    'Temperature': np.random.uniform(15, 35, size=100),
    'Humidity': np.random.uniform(30, 90, size=100)
}

df = pd.DataFrame(data)
print(df.head())

#2. Plot 'Temperature' and 'Humidity' on the Same Plot with Different y-axes

import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], 'b-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'r-', label='Humidity')
ax2.set_ylabel('Humidity', color='r')
ax2.tick_params('y', colors='r')

plt.title('Temperature and Humidity Over Time')

plt.show()

#3. Create a NumPy Array with 1000 Samples from a Normal Distribution and Plot Histogram with PDF Overlay

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Data Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, data.mean(), data.std())
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()

#4. Create a Seaborn Scatter Plot of Two Random Arrays with Color-Coded Quadrants

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)

colors = ['red' if xi > 0 and yi > 0 else 'blue' if xi < 0 and yi > 0 else 'green' if xi < 0 and yi < 0 else 'purple' for xi, yi in zip(x, y)]

df_scatter = pd.DataFrame({'x': x, 'y': y, 'color': colors})

plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', hue='color', data=df_scatter, palette=['red', 'blue', 'green', 'purple'])

plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend(title='Quadrants', loc='upper right')

plt.show()
```

```python
"""g Using Bo+eh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'8"""


from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

output_notebook()

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=4)

data = pd.DataFrame({'categories': categories, 'values': values})

source = ColumnDataSource(data)

p = figure(x_range=categories, plot_height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color='color')

hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values")
]
p.add_tools(hover)

p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

show(p)
```
